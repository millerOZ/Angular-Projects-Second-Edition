```python
#Importar funcion de librerías, crear el dataframe
from geopy.geocoders import Nominatim
from pandas.core.frame import DataFrame
from geopy.extra.rate_limiter import RateLimiter
import folium 

accidentes = pd.read_excel('Muertoss.xlsx',sheet_name='MUERTOS')
print(accidentes.head())

print(accidentes.columns)
matrizValues = accidentes
i=0
while i < len(accidentes):
  drc = accidentes.loc[i][14]
  #ciudad = accidentes.loc[i][11]
  #pais = accidentes.loc[i][3]

  direccion = drc + ", " + "Cali" + ', ' + "Colombia" #pais concatenar
  #print(direccion)

  #matrizValues.append(pd.DataFrame(direccion))
  i+=1

print(matrizValues)

dataset = pd.DataFrame(accidentes)
dataset.head(5)
```

```python

localizador = Nominatim(user_agent = 'Geocodificador')

#Hacemos un delay para evitar que el geolocalizador no niegue el acceso
geocode = RateLimiter(localizador.geocode, min_delay_seconds = 0.1)
#Creamos la localizacion y se guarda en una columna
dataset['localizacion'] = dataset['DIRECCION ' ].apply(geocode) 
print(dataset['localizacion'])
dataset.head(5)
```

```python
#Obtenemos el punto
dataset['punto'] = dataset['localizacion'].apply(lambda loc: tuple(loc.point) if loc else None)
dataset.head(5)
```

```python
#Dividimos el punto en latitud y longitud 
dataset[['lat','long','alt']] = pd.DataFrame(dataset['punto'].tolist(),index = dataset.index)
dataset.head(5)
```

```python
#Creamos mapa centrado en Cali
Mapa = folium.Map(location = [3.43722,-76.5225],zoom_start = 12)
#Recorremos el dataset adicionando cada fila al mapa
dataset.apply(lambda fila: folium.CircleMarker (location = [fila["lat"],fila["lng"]]).add_to(Mapa), axis = 1)
Mapa 

```
